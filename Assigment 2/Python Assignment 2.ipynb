## Q11
In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

### Answer
- `range(10)`: the `10` value is passed to `stop` argument, the value will be range from 0 to 9. (`0` is default value).
- `range(0, 10)`: the `0` value is passed to `start` and the `10` value is passed to `stop` argument, the value will be range from 0 to 9.
- `range(0, 10, 1)`: the `0` value will be passed to `start`, the `10` value is passed to `stop`, and `1` value is passed to `step` argument, the value will be range from 0 to 9 with the step of 1.

## Q12
